```python
import pandas as pd

excel_file_path = '/content/sample_data/Residents.xlsx'
residents_df = pd.read_excel(excel_file_path)

display(residents_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each care type
care_type_counts = residents_df['Care Type'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=care_type_counts.index, y=care_type_counts.values)
plt.title('Distribution of Residents by Care Type')
plt.xlabel('Care Type')
plt.ylabel('Number of Residents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Task
Visualize the data from "Residents.xlsx" by creating a distribution plot for age, a pie chart for gender, a bar plot for state, and a distribution plot for BMI.

## Visualize age distribution

### Subtask:
Create a distribution plot (e.g., histogram or density plot) to show the distribution of ages of the residents.


**Reasoning**:
Calculate the age of each resident and create a histogram to visualize the age distribution.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Replace '90+' with 90 in the 'Age' column and convert to numeric
residents_df['Age'] = residents_df['Age'].replace('90+', 90).astype(float)

# Create a histogram of the 'Age' column
plt.figure(figsize=(10, 6))
sns.histplot(data=residents_df, x='Age', kde=True)
plt.title('Distribution of Resident Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Visualize gender distribution

### Subtask:
Create a pie chart to show the proportion of male and female residents.


**Reasoning**:
Count the occurrences of each gender and create a pie chart to visualize the distribution, including labels and percentages.



```python
gender_counts = residents_df['Gender'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Residents by Gender')
plt.axis('equal')
plt.show()
```

## Visualize state distribution

### Subtask:
Create a visualization (e.g., bar plot) to show the distribution of residents across different states.


**Reasoning**:
Count the occurrences of each state and create a bar plot to visualize the distribution.



```python
state_counts = residents_df['State'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=state_counts.index, y=state_counts.values)
plt.title('Distribution of Residents by State')
plt.xlabel('State')
plt.ylabel('Number of Residents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Visualize bmi distribution

### Subtask:
Create a distribution plot (e.g., histogram or density plot) to show the distribution of BMI values.


**Reasoning**:
Create a distribution plot for the 'BMI' column as instructed.



```python
# Filter out rows where BMI is 0
filtered_residents_df = residents_df[residents_df['BMI'] != 0]

plt.figure(figsize=(10, 6))
sns.histplot(data=filtered_residents_df, x='BMI', stat='count') # Removed kde=True
plt.title('Distribution of Resident BMI (Excluding 0)')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```
